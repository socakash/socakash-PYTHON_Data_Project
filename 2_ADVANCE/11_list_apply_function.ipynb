```python
import pandas as pd
import matplotlib as plt
import ast 

file = pd.read_csv('C:\\Users\\akash\\Downloads\\PYTHON\\data_jobs.csv')

file.job_posted_date = pd.to_datetime(file.job_posted_date)

file['job_posted_month'] = file.job_posted_date.dt.month_name()
```

```python
def clean_list(skills):
    if pd.notna(skills):
        return(ast.literal_eval(skills))

file.job_skills = file.job_skills.apply(clean_list)
```

```python
# OR we can use lambda function for this 

file['job_skills'] = file['job_skills'].apply(
    lambda skills: ast.literal_eval(skills) if isinstance(skills, str) else skills)
```
