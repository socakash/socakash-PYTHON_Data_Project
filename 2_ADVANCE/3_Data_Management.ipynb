```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = pd.read_csv('C:\\Users\\akash\\Downloads\\PYTHON\\data_jobs.csv')
```

```python
file_new = file.copy()
```

```python
file_new.sample()
```

```python
# Table with rows = all countries , columns = all job title,  data = median salary for all jobs 

file_new.pivot_table(
    index= 'job_country', 
    columns= 'job_title_short', 
    values= 'salary_year_avg', 
    aggfunc= 'median'
    )
```

```python
# Now filter for top 5 countries by job count and 3 specific job titles 

# top 5 countries by job count
top_job_nation = file_new.job_country.value_counts().head(5).index

# Make table before filtering 
new_table = file_new.pivot_table(
    index= 'job_country', 
    columns= 'job_title_short', 
    values= 'salary_year_avg', 
    aggfunc= 'median'
    )

# now filter the table with 5 countries 
new_table = new_table.loc[top_job_nation]

# now filter the new table for 3 job titles 
job_titles = ['Business Analyst', 'Data Analyst','Data Engineer']


new_table = new_table[job_titles]

new_table.plot(kind= 'bar')
```
