# What are the most demanded skills for the top-3 most popular job roles

1- clean skills column 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import ast
import seaborn as sns

df = pd.read_csv("C:/Users/akash/Downloads/PYTHON/data_jobs.csv")

```

# Clean Job_posted_date to date time frame 
# clean job skills str to list

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df.dropna(subset= ['job_skills'], inplace= True)

df['job_skills'] = df['job_skills'].apply(lambda skills : ast.literal_eval(skills) )
```

```python
df_new = df.copy()
df_new
```

# filter large dataset for Country = USA and Role Data Analyst

```python
df_da_us = df_new[(df_new.job_country =='United States') & (df_new.job_title_short == 'Data Analyst') ]
```

# Explore Dataset check each column and values associated 

```python
df_plot = df_da_us['job_location'].value_counts().head(10).to_frame()
df_plot
```

```python
sns.set_theme (style='whitegrid')
sns.barplot(
    data= df_plot,
    x= 'count',
    y= 'job_location',
    hue= 'count',
    palette= 'light:g'
)
```

# Explore Company column 

```python
df_company = df_da_us['company_name'].value_counts().head(10).to_frame()
```

```python
sns.barplot(
    data= df_company, 
    y= 'company_name', 
    x='count', 
    hue= 'count', 
    palette= 'light:b')
```
