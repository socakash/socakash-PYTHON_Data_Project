What are the most demanded skills for the top-3 most popular job roles

1- clean skills column 
2- calc skill count based on Job_title_short 
3- plot initial finding 
4- clac skill %
5- plot final findings 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import ast
```

```python
df = pd.read_csv("C:/Users/akash/Downloads/PYTHON/data_jobs.csv")
```

```python
df_new = df.copy()
```

```python
# Clean Job_posted_date to date time frame 
# clean job skills str to list
# Filter for Canada 

df_new['job_posted_date'] = pd.to_datetime(df_new['job_posted_date'])

df_new.dropna(subset= ['job_skills'], inplace= True)

df_new['job_skills'] = df_new['job_skills'].apply(lambda skills : ast.literal_eval(skills) )


df_new
```

```python
# filter large dataset for Country = Canada and job_title = BA, DA, CE, DE, DS, SE

df_new = df_new[ df.job_title_short.isin
       ([
           'Business Analyst', 
           'Cloud Engineer', 
           'Data Analyst', 
           'Data Engineer', 
           'Data Scientist', 
           'Software Engineer'
        ])
       &
       (df_new['job_country'] =='Canada')
]

df_new
```

```python
#2- calc skill count based on Job_title_short 

# most in-demand jobs 

df_job_count = df_new.groupby('job_title_short').agg(
    job_count = ('job_title_short', 'size')
)

print(df_job_count)

df_job_count.sort_values(by='job_count', ascending= False).head(3)
```

```python
df_new.job_location.value_counts()
```

```python
# calc the skills required based on the top-3 jobs
df_new = df_new.explode(['job_skills'])

# filter table based on the top-3 jobs 

df_filter = df_new[df_new.job_title_short.isin(
    ['Data Engineer', 'Data Analyst', 'Data Scientist']
)]



```

```python
df_top_3_skills = df_filter.groupby('job_skills').agg( 'size').sort_values(ascending= False).head(3)
```

```python
df_top_3_skills = df_top_3_skills.reset_index()

df_top_3_skills
```
