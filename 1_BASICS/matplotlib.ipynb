```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
file = pd.read_csv("/content/sample_data/data_jobs.csv")
```

```python
file.info()
```

```python
file.job_posted_date = pd.to_datetime(file.job_posted_date)
```

```python
file.drop(['salary_hour_avg'], axis =1, inplace = True)
```

```python
file.drop(['job_health_insurance'], axis =1, inplace = True)
```

```python
file.dropna(subset='salary_year_avg', inplace= True)
```

```python
file['job_posted_month'] = file['job_posted_date'].dt.month
```

```python
monthly_job_count = file['job_posted_month'].value_counts()
monthly_job_count = monthly_job_count.sort_index()
```

```python
plt.plot(monthly_job_count.index, monthly_job_count)
```

```python
plt.bar(monthly_job_count.index, monthly_job_count)
```

```python
job_count = file.job_title_short.value_counts()
job_count = job_count.sort_index()

#plt.figure(figsize=(15, 8))  # Optional: set figure size
plt.xlabel('Job Title')
plt.xticks(rotation = 90)
plt.ylabel('No. of job posting')
plt.title('Job Postings by Title')
plt.bar(job_count.index, job_count)
plt.tight_layout()  # Optional: improves spacing
plt.show()
```

```python
job_count = file.job_title_short.value_counts()
job_count = job_count.sort_index()

plt.barh(job_count.index, job_count)
```

```python
# PLOT USING PANDAS
job_count.plot(kind= 'bar')

# add matplotlib to make it look nicer
plt.xlabel('Job Title')
plt.xticks(rotation = 45, ha = 'right')
plt.ylabel('No. of jobs')


plt.title('Job Postings by Title')
plt.show()
```
