```python
file['job_posted_date'] = pd.to_datetime(file['job_posted_date'])
```

```python
file['job_posted_month'] = file.job_posted_date.dt.month
```

```python
file.sort_values('job_posted_date', inplace = True)
```

```python
file.drop('job_posted_month', axis=1, inplace = True)
file.drop('salary_hour_avg', axis=1, inplace = True)
file.drop('job_health_insurance', axis=1, inplace = True)
```

```python
file.dropna(subset='salary_year_avg', axis=0, inplace= True)
```
