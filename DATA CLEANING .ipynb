```python
import pandas as pd 
df = pd.read_excel(r"C:\Users\akash\Downloads\PYTHON\Alex_course\Customer Call List.xlsx")
```

```python
df.drop_duplicates(inplace= True)
```

```python
# remove column that are not required 
df = df.drop(columns='Not_Useful_Column')
```

```python
# clean last name column 
#df.Last_Name = df.Last_Name.str.strip('...')
#df.Last_Name = df.Last_Name.str.strip('/')
#df.Last_Name = df.Last_Name.str.strip('_')

df.Last_Name = df.Last_Name.str.strip('123/_.!@#%^&*(_=+')
```

```python
# keep letter and number and remove everything else 
df.Phone_Number = df['Phone_Number'].str.replace('[^0-9]', '', regex= True)
```

```python

# convert all values to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# now make it in format 000-000-0000

df.Phone_Number = df.Phone_Number.apply(lambda y : y[0:3]+'-'+ y[3:6]+'-'+ y[6:10])
```

```python
# now remove null NA NAN and any other null values
df.Phone_Number.str.replace('[^-0-9]', '', regex= True)
df.Phone_Number = df.Phone_Number.str.replace('nan', '', regex= True)
```

```python
# NOW split the Address column and make a new - PINCODE column 

df[['Street_Address', 'City', 'Pin_Code']] = df['Address'].str.split(',', n=2, expand= True)
df
```

```python
df.drop(columns = 'Address', inplace= True)
```

```python
df['Paying Customer'] = df['Paying Customer'].replace({'YES': 'Y', 'Yes' : 'Y', 'No': 'N', 'no' : 'N'})
df['Paying Customer'] = df['Paying Customer'].replace({'Y': 'YES', 'N' : 'NO'})
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y': 'YES', 'N': 'NO'})
```

```python
df
```

```python
df = df.replace('N/a', '--')

# You cannot replace NaN you have to fill it 
df = df.fillna('--')
df
```
